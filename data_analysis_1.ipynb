List Comprehension

```python
fruits = ['apple', 'banana', "cherry", "kiwi", "mango"]
newList = []
for x in fruits:
    if "a" in x:
        newList.append(x)
        
newList
```

```python
# With List comprehension
newList = [x for x in fruits if "a" in x]
newList
```

Back to Chipotle Dataframes

```python
# clean the item_price column and transform it in a float
prices = [float(value[1:-1]) for value in chipoDF.item_price]
#prices
```

```python
# reassign the column with the cleaned prices
chipoDF.item_price = prices
```

```python
# print the Dataframe to see the new cleaned price 
chipoDF
```

```python
chipoDF_filtered = chipoDF.drop_duplicates(['item_name', 'quantity'])
#chipoDF_filtered
```

```python
chipoDF_one_product = chipoDF_filtered[chipoDF_filtered.quantity == 1]
#chipoDF_one_product
```

```python
# select only the products with quantity equals to 1
chipoDF_one_product = [chipoDF_filtered.quantity == 1]
```

```python
chipoDF_one_product[chipoDF_one_product['item_price'] >10].item_name.nunique()
```

```python
# what is the price of each item?
# print a data frame only two columns, item_name and item_price
chipoDF_filtered = chipoDF.drop_duplicates(['item_name', 'quantity'])
# select only the products with quantity equals 1
chipoDF_one_product = chipoDF_filtered[chipoDF_filtered.quantity == 1]
#select only the item_name an item price columns
price_per_item = chipoDF_one_product[['item_name', 'item_price']]
# sort the values from the most to less expensive
price_per_item.sort_values(by="item_price", ascending=False).head(10)
```
